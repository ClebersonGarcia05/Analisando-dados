```python
import pandas as pd
import plotly.express as px

table = pd.read_csv('Zoom.csv')
pd.set_option('float_format', '{:,.2f}'.format)
table['Diferença'] = table['Open'] - table['Close']

display(table[:100])
```

```python
fig = px.bar(table[:100], x='Date', y='Volume', title='Quantidade de ações vendidas entre abril e setembro de 2019', color='Volume')
fig.show()
```

```python
fig = px.sunburst(table[0:20], path=['Open', 'Close', 'Diferença'], title='Diferença de valor entre a abertura e fechamento das ações', 
color='Diferença')
fig.show()
```

```python
fig = px.line(table, x='Date', y='Open', title='Valor de abertura das ações entre 2019 e 2022')
fig.show()
```
